### Extracting brightness from multiple points (determined by clicking)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the video file path
video_file = 'video.mp4'

# Create a VideoCapture object to read the video
cap = cv2.VideoCapture(video_file)

# Check if the video file was opened successfully
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

# Create an empty list to store brightness values over time
brightness_values = []

# Create a window to display the video frame
cv2.namedWindow('Video Frame')

def get_brightness(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        # Extract the brightness value at the selected point
        pixel_value = frame[y, x]
        brightness = np.mean(pixel_value)
        brightness_values.append(brightness)

# Set the mouse callback function
cv2.setMouseCallback('Video Frame', get_brightness)

while True:
    ret, frame = cap.read()

    # Break the loop when the video ends
    if not ret:
        break

    # Display the video frame
    cv2.imshow('Video Frame', frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video file and close the OpenCV windows
cap.release()
cv2.destroyAllWindows()

# Plot the brightness values over time
plt.plot(brightness_values)
plt.xlabel('Frame Number')
plt.ylabel('Brightness Value')
plt.title('Brightness Over Time')
plt.show()
```

### Extracting brightness from a point in the video 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the video file path
video_file = 'video3.mp4'

# Create a VideoCapture object to read the video
cap = cv2.VideoCapture(video_file)

# Check if the video file was opened successfully
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

# Create an empty list to store brightness values over time
brightness_values = []

# Create a window to display the video frame
cv2.namedWindow('Video Frame')

# Create a list to store the selected point
selected_point = []

def get_brightness(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        # Store the selected point
        selected_point.clear()
        selected_point.extend([x, y])

# Set the mouse callback function
cv2.setMouseCallback('Video Frame', get_brightness)

delay=30
while True:
    ret, frame = cap.read()

    # Break the loop when the video ends
    if not ret:
        break

    # Display the video frame
    cv2.imshow('Video Frame', frame)

    if selected_point:
        # Extract the brightness value at the selected point
        x, y = selected_point
        pixel_value = frame[y, x]
        brightness = np.mean(pixel_value)
        brightness_values.append(brightness)

    # Press 'q' to exit the loop
    if cv2.waitKey(delay) & 0xFF == ord('q'):
        break

# Release the video file and close the OpenCV windows
cap.release()
cv2.destroyAllWindows()

# Plot the brightness values over time
plt.plot(brightness_values)
plt.xlabel('Frame Number')
plt.ylabel('Brightness Value')
plt.title('Brightness Over Time at the Selected Point')
plt.show()
```
